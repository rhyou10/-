```python
import requests
import pandas as pd
```

```python
url = "https://coinpan.com/index.php?mid=hot&page=1"
req = requests.get(url)
html = req.text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html,'html.parser')
```

```python
#board_list > table > tbody > tr:nth-child(6)
#board_list > table > tbody > tr:nth-child(6)

trs =  soup.select('.board_list tbody tr') #bg1 , bg2 가 나누어져있네 한번에 못받나?
#trs를 두번받아서 합치면 될까? 어차피 순번은 정해져있으니까

len(trs)
```

```python
table = []
a= 0
for i in range(1,52):
    url = "https://coinpan.com/index.php?mid=hot&page="+f'{i}'
    req = requests.get(url)
    html = req.text
    soup = BeautifulSoup(html,'html.parser')
    trs =  soup.select('.board_list tbody tr')

    for tr in trs[2:]:
        no = tr.select_one('.no .number').get_text().strip()
        notice_board = tr.select_one('.no .mid').get_text().strip()
        title = tr.select_one('.title a').get_text().strip()
        author = tr.select_one('.author').get_text().strip()
        time = tr.select_one('.time').get_text().strip()
        readed = tr.select_one('.readed').get_text().strip()
        vote = tr.select_one('.voted').get_text().strip().split(' - ')
        rec_vote = vote[0]
        rec_opp = vote[1]
        table.append([int(no), notice_board, title, author, time, int(readed), int(rec_vote), int(rec_opp)])
```

```python
df = pd.DataFrame(table,columns=['번호', '게시판', '제목', '닉네임', '등록일', '조회수', '추천', '반대'])
```
